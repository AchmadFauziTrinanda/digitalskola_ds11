```python
import pandas as pd
import numpy as np
```

# ***Load Dataset 1***

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data.csv")
df1.head()
```

# ***Load Dataset 2***

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data_2.csv")
df2.head()
```

# ***Full Outer Join***

```python
df3 = df1.merge(df2,on='employee_id',how='outer')
df3.head()
```

Check Dimension


```python
df3.shape
```

```python
df_clean = df3.loc[(df3["awards_won_y"]).isnull()]
df_clean.head()
```

```python
df_clean.shape
```

Terdapat perbedaan setelah mencari data yang "null" jumlah row berkurang dari 59858 menjadi 54022 row

# ***Mengambil Dataset 1 Saja***

```python
df_new = df_clean.iloc[:,0:14]
df_new.head()
```

#***Mengubah Nama Kolom***

```python
df_new.rename(columns = {'Unnamed: 0_x': 'Unnamed: 0',
                         'department_x':'department',
                         'region_x':'region',
                         'education_x' :'education',
                         'gender_x': 'gender',
                         'recruitment_channel_x' : 'recruitment_channel',
                         'no_of_trainings_x' : 'no_of_trainings',
                         'age_x' : 'age',
                         'previous_year_rating_x' : 'previous_year_rating',
                         'length_of_service_x' : 'length_of_service',	
                         'awards_won_x' : 'awards_won',	
                         'avg_training_score_x' : 'avg_training_score',
                         'is_promoted_x' : 'is_promoted'}).head()
```
