**Homework 3 - Introduction and Basic Dataframe (Pandas)**

image.png

In this homework, you need to create dataframe from this above table. The table has 4 rows and 11 columns. Your tasks are to create dataframe in:
- list
- tupple
- numpy
- pandas series

```python
#list
list1 = ['462809','462643','466315','461735']
list2 = ['Male','Female','Female','Male']
list3 = ['No','Yes','Yes','Yes']
list4 = [22,38,67,67]
list5 = ['No','Yes','Yes','Yes']
list6 = ['Healthcare', 'Engineer','Engineer','Lawyer']
list7 = ['1.0',' ','1.0','0.0']
list8 = ['Low','Average','Low','High']
list9 = [4.0,3.0,1.0,2.0]
list10 = ['Cat_4','Cat_4','Cat_6','Cat_6']
list11 = ['D','A','B','B']

comp = list(zip(list1, list2, list3, list4, list5, list6, list7, list8, list9, list10, list11)) 

list_dataframe = pd.DataFrame(comp, columns=['ID','Gender','Ever_Married','Age','Graduated','Profession','Work_Experience','Spending_Score','Family_Size',
                                   'Var_1','Segmentation'])
list_dataframe
```

```python
#tuple
tuple1 = ('462809','Male','No',22,'No','Healthcare','1.0','Low',4.0,'Cat_4','D')
tuple2 = ('462643','Female','Yes',38,'Yes','Engineer',' ','Average',3.0,'Cat_4','A')
tuple3 = ('466315','Female','Yes',67,'Yes','Engineer','1.0','Low',1.0,'Cat_6','B')
tuple4 = ('461735','Male','Yes',67,'Yes','Lawyer','0.0','High',2.0,'Cat_6','B')

tuple_dataframe = pd.DataFrame([tuple1, tuple2, tuple3, tuple4], columns=['ID','Gender','Ever_Married','Age','Graduated','Profession','Work_Experience',
                                                                     'Spending_Score','Family_Size','Var_1','Segmentation'])
tuple_dataframe
```

```python
#numpy
arr = np.array([['462809','Male','No',22,'No','Healthcare','1.0','Low',4.0,'Cat_4','D'],
                ['462643','Female','Yes',38,'Yes','Engineer',' ','Average',3.0,'Cat_4','A'],
                 ['466315','Female','Yes',67,'Yes','Engineer','1.0','Low',1.0,'Cat_6','B'],
                ['461735','Male','Yes',67,'Yes','Lawyer','0.0','High',2.0,'Cat_6','B']])

dataframe_arr = pd.DataFrame(arr, columns=['ID','Gender','Ever_Married','Age','Graduated','Profession','Work_Experience',
                                                                     'Spending_Score','Family_Size','Var_1','Segmentation'])
dataframe_arr
```

```python
#pandas series
series1 = (['462809','462643','466315','461735'])
series2 = (['Male','Female','Female','Male'])
series3 = (['No','Yes','Yes','Yes'])
series4 = ([22,38,67,67])
series5 = (['No','Yes','Yes','Yes'])
series6 = (['Healthcare', 'Engineer','Engineer','Lawyer'])
series7 = (['1.0',' ','1.0','0.0'])
series8 = (['Low','Average','Low','High'])
series9 = ([4.0,3.0,1.0,2.0])
series10 = (['Cat_4','Cat_4','Cat_6','Cat_6'])
series11 = (['D','A','B','B'])

dataframe_series = pd.DataFrame({'ID':series1,'Gender':series2,'Ever_Married':series3,'Age':series4,'Graduated':series5,'Profession':series6,
                          'Work_Experience':series7,'Spending_Score':series8,'Family_Size':series9,'Var_1':series10,'Segmentation':series11})
dataframe_series
```
