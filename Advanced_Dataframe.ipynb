#**Advanced Dataframe**

```python
import pandas as pd
import numpy as np

import plotly.express as px
import matplotlib.pyplot as plt
```

#**Read Data**

```python
data = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/insurance.csv")
data.head()
```

#**Lambda Function**

```python
data["bmi_category_lambda"] = data['bmi'].apply(lambda x: "Low BMI" if x < 30 else ("Medium BMI" if x < 50 else "High BMI"))
data.head()
```

#**Pivot Table**

```python
pd.pivot_table(data, index=["bmi_category_lambda"],
               aggfunc={'bmi':['max','min','mean']})
```

#**Uppercase Data in SEX, REGION and BMI CATEGORY Columns**

```python
data[['sex','region','bmi_category_lambda']] = data[['sex','region','bmi_category_lambda']].apply(lambda x:x.str.upper())
data.head()
```
